```python
overspeed_columns = ['Overspeed2_5', 'Overspeed2_510', 'Overspeed2_10',
       'Overspeed25_510', 'Overspeed25_10', 'Overspeed25_5', 'Overspeed5_5',
       'Overspeed5_510', 'Overspeed5_10', 'NightOverspeed2_5',
       'NightOverspeed2_510', 'NightOverspeed2_10', 'NightOverspeed25_5',
       'NightOverspeed25_510', 'NightOverspeed25_10', 'NightOverspeed5_5',
       'NightOverspeed5_510', 'NightOverspeed5_10']
gov_vehicle_alarm_monthly['overspeed_all'] = 0
for column in overspeed_columns:
    gov_vehicle_alarm_monthly['overspeed_all'] = gov_vehicle_alarm_monthly['overspeed_all'] + gov_vehicle_alarm_monthly[column]
```

```python
exigency_columns = ['VehicleID', 'StatMonth', 'Exigency']
exigency_data = gov_vehicle_alarm_monthly[exigency_columns]
exigency_data = pd.pivot_table(data=exigency_data, columns='StatMonth', index='VehicleID', values='Exigency').fillna(0)
exigency_data.head(10)
```

```python
overspeed_columns = ['VehicleID', 'StatMonth', 'overspeed_all']
overspeed_data = gov_vehicle_alarm_monthly[overspeed_columns]
overspeed_data = pd.pivot_table(data=overspeed_data, columns='StatMonth', index='VehicleID', values='overspeed_all').fillna(0)
overspeed_data.head(10)
```

```python
fatigue_drive_columns = ['VehicleID', 'StatMonth', 'FatigueDrive']
fatigue_drive_data = gov_vehicle_alarm_monthly[fatigue_drive_columns]
fatigue_drive_data = pd.pivot_table(data=fatigue_drive_data, columns='StatMonth', index='VehicleID', values='FatigueDrive').fillna(0)
fatigue_drive_data.head(10)
```

```python
import matplotlib.pylab as plt
def plot_data(data, vehicle_num):
    temp = data[:vehicle_num].transpose()
    temp.index = pd.to_datetime([str(d) + '01' for d in temp.index])
    temp.plot()
    plt.show()
```

```python
plot_data(exigency_data, 10)
```

```python
plot_data(overspeed_data, 10)
```

```python
plot_data(fatigue_drive_data, 20)
```
